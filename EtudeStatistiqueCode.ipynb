### Analyse statistique : test du Chi²

Pour évaluer si la répartition des espèces varie significativement selon l’âge, un **test du Chi² d’indépendance** a été appliqué à une table de contingence croisant `Species` et `age_category`.

**Résultats obtenus :**

- **Statistique de Chi²** : 65.53  
- **P-value** : 0.0469

### Interprétation

La **valeur p est légèrement inférieure au seuil de 0.05**, indiquant une **association statistiquement significative** entre les **catégories d’âge** et les **espèces bactériennes infectantes**. Ce résultat suggère que certaines espèces bactériennes sont **plus fréquemment retrouvées chez des groupes d’âge spécifiques**, ce qui peut être lié à des facteurs physiopathologiques ou à des expositions différentes selon l’âge.

Cette observation mérite une analyse plus approfondie pour comprendre quelles espèces prédominent dans chaque tranche d’âge et les implications cliniques éventuelles.

```python

from scipy.stats import chi2_contingency

# Assuming your data is stored in a DataFrame named 'data'
# Make sure your data is in the appropriate format


# Create a contingency table
contingency_table = pd.crosstab(data['age_category'], data['Species'])

# Perform the chi-squared test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Output the results
print(f"Chi-squared statistic: {chi2}")
print(f"P-value: {p}")

# Check if the result is statistically significant (using a significance level of 0.05)
if p < 0.05:
    print("There is a significant association between Age Category and Species Infection.")
else:
    print("There is no significant association between Age Category and Species Infection.")
```

### Analyse statistique : test du Chi²

Ensuite, un **test du Chi² d’indépendance** a été effectué pour déterminer s'il existe une association statistiquement significative entre **le sexe** et **les espèces bactériennes**. Une table de contingence a été construite en croisant les colonnes `gender` et `Species`, et le test a permis de calculer la statistique de Chi², les degrés de liberté et la valeur p.

**Résultats obtenus :**

- **Statistique de Chi²** : 15.82
- **Degrés de liberté** : 12
- **P-value** : 0.1998

### Interprétation

La **valeur p** étant supérieure à 0.05, nous **n'avons pas de preuve** suffisante pour rejeter l'hypothèse nulle. Cela signifie qu'il n'y a pas d'association statistiquement significative entre **le sexe** et **les espèces de bactéries infectantes** dans ce jeu de données. En d'autres termes, les différences observées dans la répartition des espèces bactériennes entre les sexes peuvent être dues au hasard et ne reflètent pas une tendance systématique.



```python
import pandas as pd
import scipy.stats as stats

# Assuming 'data' is your DataFrame
# Group data by gender and species to get the counts of each infection by gender
strain_counts = data.groupby(['gender', 'Species']).size().unstack(fill_value=0)

# Perform the chi-squared test of independence
chi2, p, dof, expected = stats.chi2_contingency(strain_counts)

# Print results
print("Chi-squared statistic:", chi2)
print("Degrees of freedom:", dof)
print("P-value:", p)
if p < 0.05:
    print("There is a significant association between Gender and Species of Infection.")
else:
    print("There is no significant association between Gender and Species of Infection.")

```

### Test du chi-deux pour la relation entre la multirésistance et l'espèce bactérienne

Cette section utilise le test du **chi-deux** pour analyser la **relation statistique** entre la **multirésistance** et les **espèces bactériennes**. Un tableau de contingence a été créé pour comparer la distribution de la multirésistance entre les différentes espèces bactériennes. Le test du chi-deux est ensuite appliqué pour vérifier si la répartition des souches selon la multirésistance dépend de l'espèce bactérienne.

Les résultats du test montrent une valeur du **chi-deux** de 30.87 et une **p-valeur** de 0.002. Comme la p-valeur est inférieure à 0.05, cela indique que la **relation entre la multirésistance et l'espèce bactérienne est statistiquement significative**. Cette analyse permet de conclure qu'il existe une association notable entre ces deux variables.


```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming your DataFrame is named 'data'
# Replace 'YourDataFrame' with the actual name of your DataFrame
# Replace 'YourSpeciesColumn' and 'YourMultiresistanceColumn' with the actual column names
contingency_table = pd.crosstab(data['Species'], data['multi_resistance_mapped'])

# Perform the chi-squared test
chi2, p, dof, _ = chi2_contingency(contingency_table)

# Output the test results
print(f"Chi-squared value: {chi2}")
print(f"P-value: {p}",dof)

# Check for significance (common threshold is 0.05)
if p < 0.05:
    print("The relationship between multiresistance and species is statistically significant.")
else:
    print("There is no significant relationship between multiresistance and species.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Melt the data for visualization
melted_data = pd.melt(df, id_vars=['family', 'Species'], var_name='variable', value_name='value')

# Create the violin plot
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))
plt.rcParams['font.family'] = 'Times New Roman'
violin_plot = sns.violinplot(
    x='family',
    y='value',
    hue='variable',
    data=melted_data,
    palette='gray'
)

# Customize the plot to make family names italic
violin_plot.set_xticklabels(
    [f"$\\it{{{label.get_text()}}}$" for label in violin_plot.get_xticklabels()],
    fontsize=12
)

# Set hatch patterns
hatch_patterns = ['-', '+', '/']
for i, patch in enumerate(violin_plot.patches):
    patch.set_hatch(hatch_patterns[i % len(hatch_patterns)])
    patch.set_edgecolor('black')

# Set labels and legend
plt.xlabel('Bacteria Group', fontweight='bold')
plt.ylabel('MIC (µg/mL)', fontweight='bold')
plt.ylim(bottom=0)
plt.legend(title='Antibiotics', title_fontsize='10', fontsize='9')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
    import matplotlib.pyplot as plt
    import numpy as np
    import pandas as pd

    # Data setup
    data = {
        "Espèce": ["Citrobacter spp.", "Escherichia coli", "Escherichia coli", "Escherichia coli", "Escherichia coli", 
                "Klebsiella pneumoniae", "Klebsiella pneumoniae", "Staphylococcus sp."],
        "Souche": ["S563*", "S705", "S389", "S595*", "S706", "S671", "S450", "S383*"],
        "R1": [22, 22, 26, 18, 23, 26, 14, 24],
        "R2": [24, 22, 24, 19, 24, 25, 13, 25],
        "R3": [22, 23, 24, 18, 25, 26, 14, 26]
    }

    # Convert data to DataFrame
    df = pd.DataFrame(data)
    df["Mean"] = df[["R1", "R2", "R3"]].mean(axis=1)
    df["Std"] = df[["R1", "R2", "R3"]].std(axis=1)

    # Unique species and colors for grouping
    species = df["Espèce"].unique()
    cmap = plt.get_cmap("tab20")  # ggplot-like color map
    colors = [cmap(i) for i in range(len(species))]
    species_color_map = dict(zip(species, colors))
    bar_positions = []
    current_position = 0

    # Set up figure
    plt.figure(figsize=(12, 8))

    # Plotting each species group
    for specie in species:
        # Filter data for the current species
        specie_df = df[df["Espèce"] == specie]
        positions = np.arange(current_position, current_position + len(specie_df))
        bar_positions.append(positions.mean())
        bars = plt.bar(
            positions, 
            specie_df["Mean"], 
            yerr=specie_df["Std"], 
            capsize=2, 
            color=species_color_map[specie], 
            edgecolor="black", 
            label=specie
        )
        
        # Add strain labels and asterisks for multiresistant strains
        for bar, souche in zip(bars, specie_df["Souche"]):
            plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 2, souche, ha="center", fontsize=11, color="black")

        current_position += len(specie_df) + 1  # Leave space between groups

    # Setting x-ticks and labels
    plt.xticks(np.hstack(bar_positions), species, fontsize=14, rotation=45)
    plt.xlabel("Espèces")
    plt.ylabel("Diamètre d'inhibition moyen (mm)")


    # Save as PDF
    plt.tight_layout()
    plt.savefig("ResultsSarra/8-inhibition_diameters.pdf")
    plt.show()
```
